```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import matplotlib.mlab as mlab
```

```python
n_sample =1000000
mu, sigma = -0.107505 , 0.3301758
log_sample = np.random.lognormal(mean=mu, sigma=sigma, size=n_sample)
```

```python
L = np.mean(log_sample)
print(L)
print(L/(1.0+L))
```

```python
# the histogram of the data
n, bins, patches = plt.hist(log_sample, 50, normed=1, facecolor='green', alpha=0.75)

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=1)

plt.xlabel('Smarts')
plt.ylabel('Probability')
# plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=80,\ \sigma=40$')
# plt.axis([40, 160, 0, 0.03])
plt.grid(True)

plt.show()
```

```python
n, bins, patches = plt.hist(log_sample, 100, facecolor='blue', alpha=0.5)
plt.show()
```

```python
log_p = map(lambda L: L/(1.0+L),log_sample)
```

```python
np.mean(log_p)
```

```python
plt.plot([prob if prob>0.5 else np.nan for prob in log_p],'r.')
plt.plot([prob if prob<0.5 else np.nan for prob in log_p],'b.')
plt.show()
```

```python
sorted_log_p = np.sort(log_p)
q_025=int(len(log_p)*0.025)
q_975=int(len(log_p)*0.975)

print(q_025,sorted_log_p[q_025])
print(q_975,sorted_log_p[q_975])
#       ,,q_975,sorted_log_p[q_975])
```
